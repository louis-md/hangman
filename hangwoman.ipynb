# Hang(wo)man

```python
import random
import re

list_of_words = [
    "apple",
    "banana",
    "orange",
    "strawberry",
    "grape",
    "blueberry",
    "melon",
    "lemon",
    "lime",
    "coconut",
    "apricot",
    "avocado",
    "blackberry",
    "cherry",
    "cranberry",
    "date",
    "fig",
    "gooseberry",
    "guava",
    "kiwi",
    "mango",
    "mulberry",
    "nectarine",
    "olive",
    "papaya",
    "peach",
    "pear",
    "pineapple",
    "plum",
    "pomegranate",
    "raisin",
    "watermelon",
]
word_to_guess = random.choice(list_of_words)
letters_to_guess = list(set(word_to_guess))
letters_guessed = []
errors_count = 0
max_errors = 10
game_won = False

def update_hidden_word(hidden):
    for i, letter in enumerate(word_to_guess):
        if letter in letters_guessed:
            hidden[i] = letter
        else:
            hidden[i] = "_"
    return hidden


hidden_word = ["_"] * len(word_to_guess)
print(
    f"Welcome to Hang(wo)man! Word to guess is {len(''.join(hidden_word))} letters long."
)

while not game_won and errors_count < max_errors:
    userInput = input('Please guess a letter: ').lower()
    if not re.fullmatch("^[a-z]{1}$", userInput):
        print("You did not type a single letter! ğŸ˜­ Please try again!")
    elif userInput in letters_guessed:
        print("You have already guessed this letter! Please try again!")
    elif userInput in letters_to_guess:
        letters_guessed.append(userInput)
        print("Congrats! ğŸ¥° You found one letter!")
        hidden_word = update_hidden_word(hidden_word)
        if '_' in hidden_word:
            print(f'Word to guess is {"".join(hidden_word)}')
        else:
            game_won = True
            print("ğŸ¥³ You won! Congratulations!")
    else:
        errors_count += 1
        if errors_count == max_errors:
            print("ğŸ˜µ You lost! Game over.")
        else:
            print(
                f"Oh no! This letter is not in the word we are looking for. You have {max_errors - errors_count} tries left. ğŸ˜¬ Maybe try again?"
            )
```
